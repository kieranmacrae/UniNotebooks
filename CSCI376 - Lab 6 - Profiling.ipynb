 # Profiling Results when Considering Parallelization via Work Items

## Results
As the trendline shows the processing time decreases rapidly as the work item size is increased. The spike at 4096 items is likely due to exceeding the devices max work items size.

```python
average_time   = [5277, 4016, 3533, 1828, 1977]
num_work_items = [256, 512, 1024, 2048, 4096]
```

```python
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
plt.plot(num_work_items, average_time)
plt.xlabel('Number of Work Items')
plt.ylabel('Average Execution Time (ns)')
plt.title('Kernel Processing Time with Increasing Parallelization')
plt.show()
```
